```python
import copy
from queue import PriorityQueue
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
#TODO
def sort(mas, n):
    for i in range(0, len(mas) - 1):
        smallest = i
        for j in range(i + 1, len(mas)):
            if mas[j] < mas[smallest]:
                smallest = j
        mas[i], mas[smallest] = mas[smallest], mas[i]
smatrix = matrix.copy()        
for i in range(0, user_m):
    sort(smatrix[i], user_n)

print(smatrix)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()
#TODO
def sort(mas, n):
    for i in range(1, user_n):
        j = i - 1
        temp = mas[i]
        while (mas[j] >= temp) and (j >= 0):
            mas[j + 1] = mas[j]
            j -= 1
        mas[j + 1] = temp

smatrix = matrix.copy()        
for i in range(0, user_m):
    sort(smatrix[i], user_n)

print(smatrix)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
#TODO
def sort(mas, n):
    for i in range(0, user_n - 1):
        for j in range(user_n - i - 1):
            if (mas[j] > mas[j + 1]):
                mas[j + 1], mas[j] = mas[j], mas[j + 1]

smatrix = matrix.copy()        
for i in range(0, user_m):
    sort(smatrix[i], user_n)

print(smatrix)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()
#TODO
def sort(mas, n):
    k = int(math.log2(n))
    interval = 2**k -1
    while interval > 0:
        for i in range(interval, n):
            temp = mas[i]
            j = i
            while j >= interval and mas[j - interval] > temp:
                mas[j] = mas[j - interval]
                j -= interval
            mas[j] = temp
        k -= 1
        interval = 2**k -1

smatrix = matrix.copy()        
for i in range(0, user_m):
    sort(smatrix[i], user_n)

print(smatrix)

print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
#TODO
def sort(nums):
    if len(nums) <= 1:
       return nums
    else:
        q = random.choice(nums)
        l_nums = [n for n in nums if n < q]
        e_nums = [q] * nums.count(q)
        b_nums = [n for n in nums if n > q]
        return sort(l_nums) + e_nums + sort(b_nums)

smatrix = matrix.copy()        
for i in range(0, user_m):
    smatrix[i] = sort(smatrix[i])

print(smatrix)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
start_time = time.time()
#TODO
max_size = 16

def sort(nums):
    if len(nums) <= max_size:
        return opimal_tournament_sort(nums) 
    else:
        return braketsize(nums)
    
def opimal_tournament_sort(nums):
    sorted_array = []
    pq = PriorityQueue(max_size)
    for num in nums:
        pq.put(num)
    while not pq.empty():
        next_item = pq.get()
        sorted_array.append(next_item)
    return sorted_array


def braketsize(nums):
    size = len(nums)
    winners = []
    losers = []
    sorted_array = []
    pq = PriorityQueue(max_size)
    while not pq.full():
        pq.put(nums.pop(0))

    while (len(nums) > 0 or not pq.empty()):
        temp = pq.get()
        if len(winners) == 0:
            winners.append(temp)
        elif temp >= winners[-1]:
            winners.append(temp)
        elif temp < winners[-1]:
            losers.append(temp)
        if not pq.full() and len(nums) > 0:
            pq.put(nums.pop(0))

    if len(losers) == 0:
        return winners
    else:
        nums = losers + winners
        return braketsize(nums)

smatrix = matrix.copy()        
for i in range(0, user_m):
    smatrix[i] = sort(smatrix[i])

print(smatrix)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
