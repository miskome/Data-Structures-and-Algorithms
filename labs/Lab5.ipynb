```python
#Кривая Коха
#Глубина рекурсии: 3
def kl(size, n):
    if n == 0:
        forward(size)
    else:
        kl(size/3, n - 1)
        left(60)
        kl(size / 3, n - 1)
        right(120)
        kl(size / 3, n - 1)
        left(60)
        kl(size / 3, n - 1)
```

```python
#Кривая Гильберта
#Глубина рекурсии: 4

#Для посроения вызовите ga(size, n), где size - длинна отрезка, n -глубина фрактала

def ga(size, n):
    if n == 1:
        setheading(180)
        forward(size)
        setheading(90)
        forward(size)
        setheading(0)
        forward(size)
    else:
        gd(size/2, n - 1)
        setheading(180)
        forward(size/(2*(n-1)))
        ga(size/2, n - 1)
        setheading(90)
        forward(size/(2*(n-1)))
        ga(size/2, n - 1)
        setheading(0)
        forward(size/(2*(n-1)))
        gb(size/2, n-1)


def gb(size, n):
    if n == 1:
        setheading(270)
        forward(size)
        setheading(0)
        forward(size)
        setheading(90)
        forward(size)
    else:
        gc(size/2, n - 1)
        setheading(270)
        forward(size/(2*(n-1)))
        gb(size/2, n - 1)
        setheading(0)
        forward(size/(2*(n-1)))
        gb(size/2, n -1)
        setheading(90)
        forward(size/(2*(n-1)))
        ga(size/2, n-1)


def gc(size, n):
    if n == 1:
        setheading(0)
        forward(size)
        setheading(270)
        forward(size)
        setheading(180)
        forward(size)
    else:
        gb(size/2, n-1)
        setheading(0)
        forward(size/(2*(n-1)))
        gc(size/2,n-1)
        setheading(270)
        forward(size/(2*(n-1)))
        gc(size/2,n-1)
        setheading(180)
        forward(size/(2*(n-1)))
        gd(size/2,n-1)

def gd(size, n):
    if n == 1:
        setheading(90)
        forward(size)
        setheading(180)
        forward(size)
        setheading(270)
        forward(size)
    else:
        ga(size/2,n-1)
        setheading(90)
        forward(size/(2*(n-1)))
        gd(size/2,n-1)
        setheading(180)
        forward(size/(2*(n-1)))
        gd(size/2,n-1)
        setheading(270)
        forward(size/(2*(n-1)))
        gc(size/2,n-1)
```

```python
#Кривая Серпинского
#Глубина рекурсии: 4

#Для построения вызовите метод krser(size, n),  где size - длинна отрезка, n -глубина фрактала
def kra(size, n):
    if n == 1:
        setheading(315)
        forward(size/2)
        left(45)
        forward(size)
        left(45)
        forward(size/2)
    else:
        kra(size, n-1)
        right(90)
        forward(size/2)
        krb(size, n-1)
        left(45)
        forward(size / 2)
        krd(size, n-1)
        right(90)
        forward(size / 2)
        kra(size, n-1)

def krb(size, n):
    if n == 1:
        setheading(225)
        forward(size/2)
        left(45)
        forward(size)
        left(45)
        forward(size/2)
    else:
        krb(size, n-1)
        right(90)
        forward(size/2)
        krc(size, n-1)
        left(45)
        forward(size / 2)
        kra(size, n-1)
        right(90)
        forward(size / 2)
        krb(size, n-1)

def krc(size, n):
    if n == 1:
        setheading(135)
        forward(size/2)
        left(45)
        forward(size)
        left(45)
        forward(size/2)
    else:
        krc(size, n-1)
        right(90)
        forward(size/2)
        krd(size, n-1)
        left(45)
        forward(size/2)
        krb(size, n-1)
        right(90)
        forward(size / 2)
        krc(size, n-1)

def krd(size, n):
    if n == 1:
        setheading(45)
        forward(size/2)
        left(45)
        forward(size)
        left(45)
        forward(size/2)
    else:
        krd(size, n-1)
        right(90)
        forward(size / 2)
        kra(size, n-1)
        left(45)
        forward(size / 2)
        krc(size, n-1)
        right(90)
        forward(size / 2)
        krd(size, n-1)

def krser(size, n):
    kra(size, n)
    right(90)
    forward(size/2)
    krb(size, n)
    right(90)
    forward(size / 2)
    krc(size, n)
    right(90)
    forward(size / 2)
    krd(size, n)
    right(90)
    forward(size / 2)
```

```python
#Салфетка Серпинского
#Глубина рекурсии: 3

def tr(size, n):
    if n == 1:
        setheading(60)
        forward(size)
        right(120)
        forward(size)
        right(120)
        forward(size)
    else:
        tr(size/2, n-1)
        setheading(60)
        forward(size/2)
        tr(size/2, n-1)
        setheading(300)
        forward(size/2)
        tr(size/2, n-1)
        forward(size/2)
```

```python
#Ковер Серпинского
#Глубина рекурсии: 9

def cover(size, n):
    if n == 1:
        pendown()
        setheading(90)
        forward(size)
        right(90)
        forward(size)
        right(90)
        forward(size)
        right(90)
        forward(size)
        penup()
    else:
        cover(size / 3, n - 1)
        right(180)
        forward(size/3)

        cover(size / 3, n - 1)
        right(180)
        forward(size / 3)

        cover(size / 3, n - 1)
        forward(2*size/3)
        right(90)
        forward(size/3)

        cover(size / 3, n - 1)
        right(180)
        forward(size / 3)

        cover(size / 3, 1)
        right(180)
        forward(size / 3)

        cover(size / 3, n - 1)
        forward(2 * size / 3)
        right(90)
        forward(size/3)

        cover(size / 3, n - 1)
        right(180)
        forward(size / 3)

        cover(size / 3, n - 1)
        right(180)
        forward(size / 3)

        cover(size / 3, n - 1)
        left(90)
        forward(2 * size / 3)
        right(90)
        forward(2 * size / 3)
```

```python
#Дерево Пифагора(классическое)
#Глубина рекурсии: 3

def pif(size, n):
    if n == 1:
        forward(size)
        right(90)
        forward(size)
        right(90)
        forward(size)
        right(90)
        forward(size)
        right(90)
    else:
        pif(size, 1)

        forward(size)
        left(45)

        pif(math.sqrt(2) * size / 2, n-1)

        right(90)
        forward(math.sqrt(2) * size / 2)

        pif(math.sqrt(2) * size / 2, n-1)

        right(90)
        forward(math.sqrt(2) * size / 2)
        right(45)
        forward(size)
        right(90)
        forward(size)
        right(90)
```
