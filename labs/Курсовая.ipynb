```python
def solutionOfnum1(countsOfTeams):
    results = []
    for count in countsOfTeams:
        temp = ''
        if count % 2 == 1:
            for i in range(count):
                for j in range(i + 1, count):
                    if j - i <= count / 2:
                        temp += '1 '
                    else:
                        temp += '-1 '
        else:
            for i in range(count):
                for j in range(i + 1, count):
                    if j - i < count / 2:
                        temp += '1 '
                    elif j - i == count / 2:
                        temp += '0 '
                    else:
                        temp += '-1 '
        results.append(temp)
    return results

def main1():
    try:
        t = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö (–æ—Ç 1 –¥–æ 100)').replace(' ', ''))
        if t < 1 or t > 100:
            print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
            return 1
        countsOfTeams = []
        for i in range(t):
            temp = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ –∫–æ–º–∞–Ω–¥ (–æ—Ç 2 –¥–æ 100)').replace(' ', ''))
            if temp < 2 or temp > 100:
                print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
                return 1
            countsOfTeams.append(temp)
    except TypeError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 2
    except ValueError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 3

    results = solutionOfnum1(countsOfTeams=countsOfTeams)
    for result in results:
        print(result)
    return 0

main1()
```

–í –ø–µ—Ä–≤–æ–π –∏ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–µ –∫–∞–∂–¥–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –∑–∞–¥–∞–Ω—ã –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ ùëõ –∏ ùëò (ùëò‚â§ùëõ<2ùëò; 1‚â§ùëò‚â§105) ‚Äî –¥–ª–∏–Ω–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ ùëé –∏ –µ–µ –º–∞–∫—Å–∏–º—É–º.

–í —Ç—Ä–µ—Ç—å–µ–º –Ω–∞–±–æ—Ä–µ, ùëé=[1,2,1] –∏ –∏–º–µ–µ—Ç 1 –∏–Ω–≤–µ—Ä—Å–∏—é. –ï—Å–ª–∏ –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º ùëù=[2,1], —Ç–æ ùëè=[ùëù[ùëé[1]],ùëù[ùëé[2]],ùëù[ùëé[3]]]=[2,1,2] –∏ —Ç–∞–∫–∂–µ –∏–º–µ–µ—Ç 1 –∏–Ω–≤–µ—Ä—Å–∏—é.

–í —á–µ—Ç–≤–µ—Ä—Ç–æ–º –Ω–∞–±–æ—Ä–µ, ùëé=[1,2,3,2], –∏ —Ç–∞–∫ –∫–∞–∫ ùëù=[1,3,2] —Ç–æ ùëè=[1,3,2,3]. –ò ùëé, –∏ ùëè –∏–º–µ—é—Ç –ø–æ 1 –∏–Ω–≤–µ—Ä—Å–∏–∏ –∏ ùëè ‚Äî –ª–µ–∫—Å–∏–∫–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞.

```python
def solutionOfnum2(data):
    results = []
    for nk in data:
        n = nk[0]
        k = nk[1]
        temp = [i for i in range(1, k - (n - k))]
        for i in range(k, k - (n - k) - 1, -1):
            temp.append(i)
        results.append(temp)
    return results

def main2():
    try:
        t = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö (–æ—Ç 1 –¥–æ 1000)').replace(' ', ''))
        if t < 1 or t > 100:
            print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
            return 1
        data = []
        for i in range(t):
            print('–í–≤–æ–¥ –Ω–∞–±–æ—Ä–∞ ', i + 1)
            n = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ n ( k <= n < 2k )').replace(' ', ''))
            k = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ k ( 1 <= k <= 105 )').replace(' ', ''))
            if k < 1 or k > 105:
                print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥(k)\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
                return 1
            elif n < k or n >= 2 * k:
                print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥(n)\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
                return 1
            data.append([n, k])
    except TypeError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 2
    except ValueError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 3

    results = solutionOfnum2(data=data)

    for result in results:
        for num in result:
            print(num, end=' ')
        print()
    return

main2()
```

### –ó–∞–¥–∞—á–∞ 3

–ö–∞–∂–¥—ã–π —Ç–µ—Å—Ç —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ –≤–≤–æ–¥–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç –æ–¥–Ω–æ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùë° (1‚â§ùë°‚â§400) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –î–∞–ª–µ–µ —Å–ª–µ–¥—É–µ—Ç –∏—Ö –æ–ø–∏—Å–∞–Ω–∏–µ.

–í –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–µ–¥–∏—Ç–µ ùëõ2 —Å—Ç—Ä–æ–∫, ùëñ-—è –∏—Ö –Ω–∏—Ö –¥–æ–ª–∂–Ω–∞ —Å–æ–¥–µ—Ä–∂–∞—Ç—å ùëéùëñ –∏ ùëèùëñ ‚Äî —ç–ª–µ–º–µ–Ω—Ç—ã ùëñ-–π –ø–∞—Ä—ã.

–ï—Å–ª–∏ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–µ—à–µ–Ω–∏–π, –≤—ã–≤–µ–¥–∏—Ç–µ –ª—é–±–æ–µ. –ü–æ—Ä—è–¥–æ–∫ –ø–∞—Ä –∏ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –ø–∞—Ä–∞—Ö –Ω–µ –∏–º–µ–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è.

–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ–°–∫–æ–ø–∏—Ä–æ–≤–∞—Ç—å\

–í –ø–µ—Ä–≤–æ–º –Ω–∞–±–æ—Ä–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (0&3)+(1&2)=0.

–í–æ –≤—Ç–æ—Ä–æ–º –Ω–∞–±–æ—Ä–µ (0&2)+(1&3)=1.

–í —Ç—Ä–µ—Ç—å–µ–º –Ω–∞–±–æ—Ä–µ (0&1)+(2&3)=2.

–í —á–µ—Ç–≤–µ—Ä—Ç–æ–º –Ω–∞–±–æ—Ä–µ —Ä–µ—à–µ–Ω–∏—è –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç.

```python
def solutionOfnum3(data):
    results = []
    for nk in data:
        temp = ''
        n = nk[0]
        k = nk[1]
        mas = [i for i in range(0, n)]
        if k == 0:
            while len(mas) > 0:
                temp += str(mas[0]) + ' ' + str(mas[-1]) + '\n'
                mas.pop(0)
                mas.pop(-1)
        elif k < n - 1:
            temp += str(mas[0]) + ' ' + str(mas[n - 1 - k]) + '\n'
            temp += str(mas[k]) + ' ' + str(mas[n - 1]) + '\n'
            mas.pop(0)
            mas.pop(k - 1)
            mas.pop(n - k - 3)
            mas.pop(n - 4)
            while len(mas) > 0:
                temp += str(mas[0]) + ' ' + str(mas[-1]) + '\n'
                mas.pop(0)
                mas.pop(-1)
        else:
            if n > 4:
                temp += str(mas[n - 1]) + ' ' + str(mas[n - 2]) + '\n'
                temp += str(mas[1]) + ' ' + str(mas[n - 3]) + '\n'
                temp += str(mas[0]) + ' ' + str(mas[2]) + '\n'
                mas.pop(0)
                mas.pop(0)
                mas.pop(0)
                mas.pop(-1)
                mas.pop(-1)
                mas.pop(-1)
                while len(mas) > 0:
                    temp += str(mas[0]) + ' ' + str(mas[-1]) + '\n'
                    mas.pop(0)
                    mas.pop(-1)
            else:
                temp = '-1'
        results.append(temp)
    return results

def main3():
    try:
        t = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ t –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö (1 <= t <= 400)').replace(' ', ''))
        if t < 1 or t > 400:
            print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
            return 1
        data = []
        for i in range(t):
            print('–í–≤–æ–¥ –Ω–∞–±–æ—Ä–∞ ', i + 1)
            n = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ n (n - —Å—Ç–µ–ø–µ–Ω—å –¥–≤–æ–π–∫–∏)').replace(' ', ''))
            k = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ k (0 <= k <= n-1)').replace(' ', ''))
            if n & (n - 1) != 0 or k < 0 or k > n - 1:
                print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
                return 1
            data.append([n, k])
    except TypeError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 2
    except ValueError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 3
    

    results = solutionOfnum3(data=data)
    for result in results:
        print(result)
    return 

main3()
```

```python
def solutionOfnum5(data):
    results = []

    for ls in data:
        temp = ''
        len = ls[0]
        string = ls[1]
        ind = string.find('0')
        if ind == -1:
            temp += '1 ' + str(len - 1) + ' 2 ' + str(len)
        elif ind < int(len / 2):
            temp += str(ind + 1) + ' ' + str(len) + ' ' + str(ind + 2) + ' ' + str(len)
        elif ind == int(len / 2) and len % 2 == 1:
            temp += '1 ' + str(len) + ' ' + str(ind + 1) + ' ' + str(len)
        else:
            temp += '1 ' + str(ind) + ' 2 ' + str(ind)
        results.append(temp)
    return results

def main5():
    try:
        n = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ t, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö  (1 <= t <= 103)').replace(' ', ''))
        if n < 1 or n > 103:
            print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
            return 1
        data = []
        for i in range(n):
            length = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–ª–µ—Ü(–æ—Ç 2 –¥–æ 2*104)').replace(' ', ''))
            if length < 2 or length > 208:
                print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
                return 1
            rings = input('–í–≤–µ–¥–∏—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∫–æ–ª–µ—Ü(–∏–∑ 0 –∏ 1)').replace(' ', '')
            if rings.count('0') + rings.count('1') != length:
                print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
                return 1
            data.append([length, rings])
    except TypeError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 2
    except ValueError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 3


    results = solutionOfnum5(data=data)

    for result in results:
        print(result)

main5()
```

```python
import math

def solutionOfnum6(data):
    def find(mas, start, value):
        ind = start
        for i in range(start, len(mas)):
            if mas[i] == value:
                return ind
            elif mas[i] > value:
                return ind
            ind += 1
        return ind

    results = []

    for la in data:
        ln = la[0]
        a = la[1]
        a.sort()
        indl = 0
        count = 0
        while indl + 2 < ln:
            indr = find(a, indl, a[indl] + 2)
            if indr == ln:
                indr -= 1
            if a[indr] - a[indl] <= 2 and indr - indl >= 2:
                count += int(math.factorial(indr - indl + 1) / (math.factorial(3) * math.factorial(indr - indl - 2)))
                if indl != 0 and indr - indl - 1 >= 2:
                    count -= int(math.factorial(indr - indl) / (math.factorial(3) * math.factorial(indr - indl - 3)))
            indl = find(a, indl, a[indl] + 1)
        results.append(count)
    return results

def main6():
    try:
        t = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ t, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö  (1 <= t <= 103)').replace(' ', ''))
        if t < 1 or t > 103:
            print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
            return 1
        data = []
        for i in range(t):
            n = int(input('–í–≤–µ–¥–∏—Ç–µ –¥–ª–∏–Ω–Ω—É –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏( 1 <= n <= 2*105)').replace(' ', ''))
            if n < 1 or n > 210:
                print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
                return 1
            a = [int(x) for x in input('–í–≤–µ–¥–∏—Ç–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∞').split()]
            if len(a) != n:
                print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
                return 1
            data.append([n, a])
    except TypeError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 2
    except ValueError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 3


    results = solutionOfnum6(data=data)

    for result in results:
        print(result)

main6()
```

```python
def solutionOfnum10(data):
    a = 0
    g = 0
    res = ''
    for pack in data:
        f = pack[0]
        s = pack[1]
        if a + f - g < 500:
            res += 'A'
            a += f
        else:
            res += 'G'
            g += s

    return res

def main10():
    try:
        n = int(input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ n (1 <= n <= 106)').replace(' ', ''))
        if n < 1 or n > 106:
            print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
            return 1
        data = []
        for i in range(n):
            pack = [int(x) for x in input('–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–∞ ai –∏ gi ( 0 <= ai, gi <= 1000; ai + gi = 1000 )').split()]
            if pack[0] < 0 or pack[1] < 0 or pack[0] > 1000 or pack[1] > 1000 or pack[0] + pack[1] != 1000:
                print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
                return 1
            data.append(pack)
    except TypeError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 2
    except ValueError:
        print('–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥\n–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã')
        return 3

    result = solutionOfnum10(data=data)

    print(result)

main10()
```
